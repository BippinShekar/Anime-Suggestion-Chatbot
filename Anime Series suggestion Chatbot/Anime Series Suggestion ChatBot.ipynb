```python
import random
```

```python
#Returning lsits of top anime in queried genre
def find_tops(user_input,accessing): #works sucesfully
    if 'top 5' in user_input:
        return accessing[:5]
    elif 'top 10' in user_input:
        return accessing[:10]
    elif 'top 15' in user_input:
        return accessing[:15]
    elif 'top 20' in user_input:
        return accessing
```

```python
#base queries answering function.
#This can handle all the genre and sub_genre types present in the data and give required answer for a query. - Works Perf
def in_data(user_input,data):
    user_input.lower()
    for i in data:
        if i.lower() in user_input:
            #to remove confusion caused by shonen and sports sub-genres
            if i.lower()=='shonen':
                if 'sports' in user_input:
                    accessing = data['Sports Shonen']
                elif 'battle' in user_input:
                    accessing = data['Battle Shonen']
                else:
                    accessing = data['Shonen']
            elif i.lower() == 'sports':
                if 'esports' in user_input:
                    accessing = data['Esports']
                elif 'motorsports' in user_input:
                    accessing = data['Motorsports']
                else:
                    accessing = data['Sports']
            else:
                accessing = data[i].values
            #returning answers to queries bassed on common keywords.
            bests = ['best','greatest','top rated','top','highest rated','amazing','most amazing','fabulous']
            found_bests = any(item in user_input for item in bests)
            tops = ['5', '10','15','20']
            found_tops = any(item in user_input for item in tops)
            if found_bests and found_tops:
                k = find_tops(user_input,accessing)
                return k
            elif found_bests:
                return accessing[0]
            suggests = ['suggest','tell','recommend','ideas','idea','clue']
            found_suggests = any(item in user_input for item in suggests)
            if found_suggests:
                return random.choice(accessing)
            else:
                return random.choice(accessing)
            
            
        #if the given query does not contain any of the mentioned genres, pass it through a different function.
    else:
        return not_in_columns(user_input,data)
```

```python
#if user_input is not similar to any columns, i.e any other key word is used,considering only major and popular categories. 
def not_in_columns(user_input,data): #Works succesfully
    fights = ['fight','fights','fighting','battle','duels','epic fights','action']
    found_fights = any(item in user_input for item in fights)
    if found_fights:
        accessing = data['Battle Shonen'].values
        if 'suggest' in user_input:
            return random.choice(accessing)
        if 'best' in user_input:
            return accessing[0]
    #entering sports shonen territory - considering answers for only specific queries(popular).
    if 'volleyball' in user_input:
        accessing = data['Sports Shonen'].values
        return (accessing[0], 'Few anime communities and fans even consider this as the best sports anime of all time.')
    elif 'basketball' in user_input:
        accessing = data['Sports Shonen'].values
        return accessing[1]
    elif 'baseball' in user_input:
        accessing = data['Sports Shonen'].values
        return accessing[2]
    elif 'story' or 'plot' in user_input:
        return single_answers(user_input,data)
```

```python
def single_answers(user_input,data): #Works Sucsefully
    if 'animation' in user_input:
        return animation(user_input,data)
    if 'best' in user_input:
        accessing = data['Seinen'].values
        return accessing[2]
    elif 'sad' in user_input:
        accessing = data['Romance'].values
        return accessing[1]
    elif 'amazing' in user_input:
        accessing = data['Gore'].values
        return accessing[1]
    elif 'brutal' in user_input:
        accessing = data['Gore'].values
        return accessing[2]
    feels = ['uncomfortable','laugh','goofy','cry','hearted']
    found_feels = any(item in user_input for item in feels)
    if found_feels:
        return feelings(user_input,data)
    else:
        return random.choice(data['Seinen'].values)
```

```python
def animation(user_input,data): #Works succesfully
    if 'animation' in user_input:
        if 'aesthetic' in user_input:
            accessing = data['Shonen'].values
            return accessing[14]
        if 'best' in user_input:
            accessing = data['Fantasy'].values
            return accessing[8]
```

```python
def feelings(user_input,data): #Work succesfully
    if 'uncomfortable' in user_input:
        accessing = data['Science Fiction'].values
        return accessing[11]
    if 'laugh' in user_input:
        accessing = data['Comedy'].values
        return accessing[1]
    if 'goofy' in user_input:
        accessing = data['Comedy'].values
        return accessing[0]
    if 'cry' in user_input:
        accessing = data['Romance'].values
        return accessing[1]
    if 'hearted' in user_input:
        accessing = data['Romantic Comedy'].values
        return accessing[1]
    else:
        return Last_Resort(data)
```

```python
#Last resort function. If the user does not mention any specific genre of emotion and jus asks for a anime suggestion, this function will kick in.
def Last_resort(data): #Works Sucessfully
    huge_set = []
    for i in data:
        accessing = data[i].values
        for i in accessing:
            huge_set.append(i)
    return random.choice(huge_set)
```
